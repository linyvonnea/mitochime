```python
import pandas as pd
import matplotlib.pyplot as plt

metrics = pd.read_csv("../reports/metrics_pair_noq/metrics_summary.tsv", sep="\t")
metrics
```

```python
metrics_sorted = metrics.sort_values("test_f1", ascending=False).reset_index(drop=True)
metrics_sorted
```

```python
plt.figure(figsize=(7, 4))
plt.barh(metrics_sorted["model"], metrics_sorted["test_f1"])
plt.xlabel("Test F1")
plt.title("Baseline models (PAIR_noq)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
metrics_auc = metrics.sort_values("test_roc_auc", ascending=False).reset_index(drop=True)

plt.figure(figsize=(7, 4))
plt.barh(metrics_auc["model"], metrics_auc["test_roc_auc"])
plt.xlabel("Test ROC-AUC")
plt.title("Baseline ROC-AUC (PAIR_noq)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
