```python
import pandas as pd
df = pd.read_csv("../../data/processed/PAIR_train_noq.tsv", sep="\t")

print("rows:", len(df))
print("unique read_id:", df["read_id"].nunique())
print("unique ref_name:", df["ref_name"].nunique())

# show how many distinct clean intervals you actually have (roughly)
clean = df[df["read_id"].str.startswith("NC_")]
clean_key = clean["read_id"].str.extract(r"^(NC_[^_]+)_(\d+)_(\d+)", expand=False)
print("unique clean intervals:", clean_key.dropna().drop_duplicates().shape[0])

chim = df[df["read_id"].str.startswith("chimera_")]
chim_key = chim["read_id"].str.extract(r"A(\d+-\d+)_B(\d+-\d+)_MH(\d+)", expand=False)
print("unique chimera recipes:", chim_key.dropna().drop_duplicates().shape[0])
```
